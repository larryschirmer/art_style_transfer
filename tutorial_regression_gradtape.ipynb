```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
class Model:
    def __init__(self):
        self.w = tf.Variable(10.0)
        self.b = tf.Variable(-5.0)
    
    def __call__(self, inputs):
        return self.w*inputs + self.b
```

```python
def compute_loss(y_true, y_pred):
    return tf.reduce_mean(tf.square(y_true - y_pred))
```

```python
model = Model()
```

```python
TRUE_w = 3.0
TRUE_b = 2.0

num_examples = 1_000
inputs = tf.random.normal(shape=[num_examples])
```

```python
plt.plot(inputs)
```

```python
outputs = TRUE_w * inputs + TRUE_b
```

```python
plt.plot(outputs)
```

```python
plt.scatter(inputs, outputs)
```

```python
outputs = outputs + tf.random.normal(shape=[num_examples])
```

```python
plt.plot(outputs)
```

```python
plt.scatter(inputs, outputs)
```

```python
lr = 0.1
epochs = 30
```

```python
for epoch in range(epochs):
    with tf.GradientTape() as tape:
        tape.watch([model.w, model.b])
        loss = compute_loss(outputs, model(inputs))

    dw, db = tape.gradient(loss, [model.w, model.b])
    model.w = tf.math.subtract(model.w,lr*dw)
    model.b = tf.math.subtract(model.b,lr*db)

    print("==> epoch: {:d}: w {:.2f} b {:.2f} loss {:.2}".format(epoch, model.w.numpy(), model.b.numpy(), loss))
```

```python
pred_outputs = model(inputs)
```

```python
plt.plot(inputs, pred_outputs)
plt.scatter(inputs, outputs, c='b')
```
